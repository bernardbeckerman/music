```python
%matplotlib inline
import matplotlib
import seaborn as sns
matplotlib.rcParams['savefig.dpi'] = 144
import librosa
import IPython.display
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.grid_search import GridSearchCV
```

```python
#%%timeit
def calc_features(mono):
    window = int(0.1*fs)
    advance = int(0.1*fs)
    zcs = []
    rmses = []
    rmse_del = []
    zc_del = []
    for i in range((len(mono)-window)/advance):
        #obtain signal over window
        isig = mono[i * advance : i * advance + window]

        #rms energy
        rmse = np.sqrt(np.sum(np.square(isig)))
    
        #zero-crossing
        pos = isig > 0
        zc  = (pos[:-1] & ~pos[1:]).nonzero()[0]

        #calculate dels
        if zcs:
            rmse_del.append(rmses[-1]-rmse)
            zc_del.append(zcs[-1]-len(zc))  
        rmses.append(rmse)
        zcs.append(len(zc))

    return [np.mean(zcs),np.std(zcs),np.mean(rmses),np.std(rmses),np.sqrt(np.sum(np.square(zc_del))),np.sqrt(np.sum(np.square(rmse_del)))]
```

```python
import os
all_features = []
for fname in sorted(os.listdir('data/train/')):
    print fname
    mono, fs = librosa.load('data/train/' + fname, sr = 44100)
    cf =  calc_features(mono)
    cf.append(fname)
    all_features.append(cf)
feat_df = pd.DataFrame(all_features, columns=['zc_mean','zc_std','rmse_mean','rmse_std','zc_del','rmse_del','filename'])
df_genre = pd.read_csv('data/music_train_labels.csv')
```

```python
param_grid = {"max_depth": [3, None],
              "max_features": [1, 3, 10],
              "min_samples_split": [1, 3, 10],
              "min_samples_leaf": [1, 3, 10],
              "bootstrap": [True, False],
              "criterion": ["gini", "entropy"]}
#rfc = GridSearchCV(RandomForestClassifier(n_estimators=100), param_grid=param_grid)
rfc = RandomForestClassifier(n_estimators=100)
print 'starting fit'
rfc.fit(feat_df.drop('filename',1),df_genre.drop('filename',1))
print 'fit complete'
```

```python
all_features_test = []
for fname in sorted(os.listdir('data/feature_extraction_test/')):
    print fname
    mono, fs = librosa.load('data/feature_extraction_test/' + fname, sr = 44100)
    cf =  calc_features(mono)
    cf.append(fname)
    all_features_test.append(cf)
feat_df_test = pd.DataFrame(all_features_test, columns=['zc_mean','zc_std','rmse_mean','rmse_std','zc_del','rmse_del','filename'])
```
